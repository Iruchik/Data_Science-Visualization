13. Проаналізуємо розподіл цін на товари в різних продуктових лінійках. Наша ціль - знайти відповідь на питання: в якій продуктовій лінійці найбільше дорогих товарів (90-100 у.о.)?

Для початку побудуємо візуалізацію і зробимо це кількома способами. 


  13.1.  На одній фігурі відобразимо гістограму розподілу цін на товари для кожної продуктової лінійки. Тобто на одній фігурі кількість гістограм рівна кількості продуктових лінійок.

```python
# для того щоб оцінити ліміти і порахувати кількість бінів
df["Unit price"].min(), df["Unit price"].max()
```

```python
fig = df.hist("Unit price",by="Product line",layout=(2, 3), edgecolor = "black", bins=range(10, 110, 10), sharey=True, color="#4284F3")
[x.title.set_size(11) for x in fig.ravel()]
plt.tight_layout()
```

13.2. А тепер відобразимо гістограми з однаковою кількість бінів на одному графіку. Для зручності використаємо прозорість alpha не більше 0.4.

```python
plt.title('Розподіл цін для продуктових ліній')
for pl in df["Product line"].unique():
  plt.hist(df[df["Product line"]==pl]["Unit price"], alpha=0.4, bins=range(10, 110, 10), label=pl)
plt.legend(loc='lower left')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.tight_layout()
```
